```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('chypoxia_expression_all.txt', sep='\t')
df.head()
```

```python
df.describe()
```

```python
HN = pd.read_csv('HN-pair.csv', sep=',')
HN.head()
```

```python
HNratio = pd.DataFrame()
HNratio["GeneName"] = df["GeneName"] 
for i in range(len(HN)):
    h = str(HN.iloc[i,0])
    n = str(HN.iloc[i,1])
#    c = str(h) +"/"+str(n)
    HNratio[h] = (df.loc[:,h] + 1)/(df.loc[:,n] +1)
    HNratio.append(HNratio)
HNratio.head()
```

```python
HNratio_ = HNratio.set_index('GeneName')

HNratiolog2 = HNratio_.apply(np.log2)
HNratiolog2 = HNratiolog2.reset_index()
HNratiolog2.head()
```

```python
HNratiolog2.describe()
```

```python
HNratiolog2.set_index('GeneName').to_csv("HN-ratio_log2.csv", sep=",")
```

```python
HNratio.set_index('GeneName').to_csv("HN-ratio.csv", sep=",")
```
